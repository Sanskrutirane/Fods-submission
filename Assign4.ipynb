```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plot
```

```python
import seaborn as sns
```

```python
# Parameters for the normal distribution
mu, sigma = 100, 4
data = np.random.normal(mu, sigma, 100000)

# Create the histogram with density plot using seaborn
plt.figure(figsize=(8, 6))
sns.histplot(data, bins=30, kde=True, stat='probability', color='blue')

# Set labels, title, and ticks
plt.xlabel('Value')
plt.ylabel('Probability')
plt.title("Normal Distribution of Values (Mean=100, Std Dev=4)")
plt.xticks(range(80, 120, 5))

# Show the plot
plt.show()
```

```python
import statsmodels.api as sm

# Parameters for the normal distribution
mu, sigma = 100, 4
data = np.random.normal(mu, sigma, 100000)

# QQ-Plot
fig = sm.qqplot(data, line='s')
plt.title('QQ-Plot of Normally Distributed Data (Mean=100, Std Dev=4)')
plt.show()
```
